Assuming that the general idea behind the flowchart above is correct, **design and write a code that implements the flowchart**.

Things to remember:
* Each decision-making step has a consequence: it leads to the next decision-making step OR to a conclusion that needs to be shared with the clinician
* When developing the code, it may help to use __incremental development__
* The code must be interactive, implying that the code must ask the clinician specific questions (e.g., "Is the person younger than 2 years of age?") and, depending on the answer, the code must proceed to one of two subsequent steps
* A clinician may enter unexpected answers. Such answers must be discarded, and the question must be asked again.
* When dealing with unexpected answers, some may simply be given in lowercase or uppercase letters: these are easy to deal with. Other unexpected answers (e.g., "negative" when "no" was expected) must be discarded, and the question must be asked again.
* Make sure that you include comments and docstrings in your code; otherwise, I won't know what you are trying to do

The code should be included in a single cell

```python
# Enter your code below
def ask(question , valid_answers):
  """
  Ask the user a question repeatedly until a valid answer is entered.

  Parameters:
       question (str): The prompt text shown to the user.
       valid_answers (dict) : A dictionary mapping acceptable input to normalised values
       (e.g., {"yes": True ,"no": False})

  Returns:
      The normalized value from valid_answers.
  """
  while True:
    user_input = input(question).strip().lower()
    if user_input in valid_answers:
      return valid_answers[user_input]
    else:
      print("Invalid answer.Please try again.\n")

def main():
  """
    Example clinical decision -making program based on a simple flowchart.
    This code demonstrates:
    -interactive questioning
    -input validation
    - branching logic based on answer
    - comments and docstrings
    - structure required for the TB
    """

  print("=== Fever Diagnosis Example (Flowchart-based) ===\n")

  # Step 1 : Does the patient have fever?
  fever = ask (
          "Does the patient have a fever? (yes/no):" ,
          {"yes": True , "no" : False }
          )
  if not fever:
    print("\nConclusion:Fever absent.Advise routine monitoring.")
    return

  # Step 2: Is the temperature above 39 C?
  high_temp = ask(
      "Is the temperature above 39 c? (yes/no):",
      {"yes":True, "no":False}
  )

  if high_temp:
    print("\nConclusion : High fever.Recommend immediate clinical assessment.")
    return

  # Step 3 : Have u ever had tb?
  had_tb = ask(
      "Have u ever had tuberculosis? (yes/no):",
      {"yes":True , "no":False}
  )
  if had_tb:
    print("\nConclusion:Possible serious condition.Recommend urgent medical evalution")



  # Step 4 : Do u have pain in joints or back
  pain = ask(
      "Do u have pain in joints or back? (yes/no):",
      {"yes":True , "no" : False}

  )
  if pain :
    print("\nConclusion: Joint or back pain reported .Monitor if symptoms persist")

  # Step 3 : Is a cough present ?
  cough = ask(
      "Is the patient coughing? (yes/no): ",
      {"yes" : True , "no" : False}
  )

  if cough :
      # Ask what type of cough(blood or phlegm)
      cough_type = ask(
          "Is the cough with blood or phlegm? (blood/phlegm):",
          {"blood":"blood","phlegm": "phlegm"}

      )
      if cough_type == "blood":
        print("\nConclusion:Possible serious condition.Recommend urgent medical evaluation.")

      else:
        print("\nConclusion:Cough with phlegm.Suggest chest examination.")

  else:
    print("\nConclusion :Mild fever without cough.Suggest hydration and rest.")

main()
```

```python
from graphviz import Digraph
from IPython.display import Image , display
import os

def create_flowchart():
  """
  Example flowchart using generic TB-style patient data.
  This is NOT the assignment logic.
  """

  flow = Digraph("TB_Flowchart_Example", format = "png")
  flow.attr(rankdir="TB", size = "35,15" , fontsize = "24")
  flow.node_attr.update(fontsize = "32",fontname ="Helvetica",style ="bold",margin = "0.5,0.3",penwidth ="2")
  flow.edge_attr.update(fontsize ="28",fontname = "Helvetica",style = "bold",color ="black")

  # Start & End

  flow.node("START","Start",shape = "oval", style = "filled" , fillcolor ="gold")
  flow.node("END", "End" , shape = "oval" , style = "filled", fillcolor = "gold")

  # Decision Nodes
  flow.node("A", "Does the patient have a fever?", shape = "diamond", style="filled" , fillcolor="lightblue")
  flow.node("B", "Is the temperature\nabove 39 C ?", shape = "diamond",style="filled" , fillcolor="lightblue")
  flow.node("C", " Does the patient had Tb earlier?" , shape = "diamond",style="filled" , fillcolor= "lightblue")
  flow.node("D", " Does the patient have pain in joints or back?" , shape = "diamond",style="filled" , fillcolor= "lightblue")
  flow.node("E", "Is the patient coughing?" , shape = "diamond",style="filled" , fillcolor= "lightblue")
  flow.node("F","Cough with blood or phlegm?" , shape = "diamond",style="filled" , fillcolor= "lightblue")

  # Process/Conclusion Nodes
  flow.node("G","Fever absent.\nRoutine monitoring.",shape = "rectangle", style = "filled",fillcolor="lightgreen")
  flow.node("H","High fever.\nImmediate clinical assessment.", shape = "rectangle" , style = "filled",fillcolor="lightgreen")
  flow.node("I", "Note:\nPrevious Tb history .\n May required screening." , shape = "rectangle", style = "filled",fillcolor="lightgreen")
  flow.node("J", "Note:\nJoint/back pain.\nMonitor symptoms." , shape = "rectangle", style = "filled",fillcolor="lightgreen")
  flow.node("K","Mild fever without cough.\nHydration and rest.",shape = "rectangle", style = "filled",fillcolor="lightgreen")
  flow.node("L","Possible serious condition.\nUrgent evalution.",shape = "rectangle", style = "filled",fillcolor="lightgreen")
  flow.node("M","Cough with phlegm.\nChest exam suggested." , shape = "rectangle", style = "filled",fillcolor="lightgreen")

  # Flow connection
  flow.edge("START","A")

  flow.edge("A" , "G" , label = "No",fontsize="28")
  flow.edge("A" ,"C" , label = "Yes",fontsize="28") # If fever is present proceed to tb question

  flow.edge("G" , "END")

  flow.edge("C" ,"I" ,label = "Yes",fontsize="28")  # Yes = had tb
  flow.edge("C" , "D" , label = "No",fontsize="28") # No = Continue

  flow.edge("I","D", "END")

  # Pain in joint/back
  flow.edge("D" ,"J"  , label = "Yes",fontsize="28")
  flow.edge("D" ,"B" , label = "No",fontsize="28")

  flow.edge("J","B","END") # Continue after the pain note
  # Temperature above 39 C
  flow.edge("B","H",label ="Yes",fontsize="28")
  flow.edge("B","E", label = "No",fontsize="28")

  flow.edge("H", "END")
  # Cough ?
  flow.edge("E","F",label = "Yes",fontsize="28")
  flow.edge("E","K",label ="No",fontsize="28")

  flow.edge("K" , "END")

  # Type of cough
  flow.edge("F","L",label = "Blood",fontsize="28")
  flow.edge("F","M",label = "Phlegm",fontsize="28")

  flow.edge("L" , "END")
  flow.edge("M","END")




 # Render the file
  filepath = flow.render("fever_diagnosis_flowchart", view = True)
  print("Flowchart saved to:",filepath)

# Display inline
if os.path.exists ("fever_diagnosis_flowchart.png"):
  display(Image(filename = "fever_diagnosis_flowchart.png"))

create_flowchart()

```
